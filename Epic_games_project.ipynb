# 1  Epic Games wants to identify the top 2 open critic rating based on the average price. Clean the Price column by removing the rupee symbol from it and store the values in a new column named as ‘Price_Clean’. Then group the data based on average price and ‘OpenCritic Rating’ to find the top 2 open critic ratings.


```python
# Remove rupees  characters and clean the Price column and renamed as Price clean
df["Price_Clean"] = df["price"].str.replace("₹","").str.replace(",","").str.replace("Free","0").astype(float)
```

```python


df["date release"] = pd.to_datetime(df["date release"])

```

```python


df.drop(columns=["game_link","available achievements","available xp"], axis=1, inplace=True)

```

```python
# Drop null values from the 'OpenCritic Rating' column
df = df.dropna(subset=['OpenCritic Rating'])
```

```python
# Map string ratings to integer values
rating_mapping = {'Strong': 4, 'Mighty': 3, 'Fair': 2, 'Weak': 1}
df['OpenCritic Rating_Integer'] = df['OpenCritic Rating'].map(rating_mapping)
```

```python
# Group the data by 'Price_Clean' and calculate the mean of 'Rating_Integer'
grouped_data = df.groupby('Price_Clean')['OpenCritic Rating_Integer'].mean()
```

```python
# Find the top 2 OpenCritic ratings
top_2_ratings = grouped_data.nlargest(2)
```

# 4 Your job is to extract the major genre for each game and store it in a new column and name the column as “Major Genre”. (Hint: All the genre name starts with uppercase). 

	•	The ‘date release’ column consist of the release date for the games. Using the column identify the average ‘Critics Recommend’ rating of a game releasing on Friday vs average ‘Critics Recommend’ rating of a game releasing on Sunday. 

	•	The ‘features’ column consists of the game type. From the column filter out only ‘Single Player’, ‘Multiplayer’ and ‘Co-op’ and store this in a new dataframe called as ‘features_filtered’. From this new dataframe, find the average ‘memory’ required for each of these three different features.

```python
# Remove '%' symbol from 'Critics Recommend' column
df['Critics Recommend'] = df['Critics Recommend'].str.replace('%', '')

print(df)
```

```python
# Function to clean and standardize memory values
def clean_memory(memory):
    try:
        # Convert to lowercase and remove non-alphanumeric characters
        memory = memory.lower().replace(' ', '').replace('ram', '')
        # Handle TBD and NAN
        if memory in ['tbd', 'nan']:
            return np.nan
        # Extract numeric value
        numeric_value = float(''.join(filter(str.isdigit, memory)))
        # Convert to GB
        if 'gb' in memory:
            return numeric_value
        elif 'mb' in memory:
            return numeric_value / 1024
        else:
            return np.nan  # Unknown unit
    except:
        return np.nan  # Handle any other errors
# Clean memory column and create memory_cleaned column
df['memory_cleaned'] = df['memory'].apply(clean_memory)
```

```python

# Extracting major genre and storing it in a new column
df['Major Genre'] = df['genres'].str.split(',').str[0]
```

```python
# Converting 'date release' column to datetime format
df['date release'] = pd.to_datetime(df['date release'])
```

```python
# Calculating average 'Critics Recommend' rating for games releasing on Friday vs. Sunday
friday_avg_rating = df[df['date release'].dt.dayofweek == 4]['Critics Recommend'].mean()
sunday_avg_rating = df[df['date release'].dt.dayofweek == 6]['Critics Recommend'].mean()
```

```python
# Filtering out 'Single Player', 'Multiplayer', and 'Co-op' from the 'features' column
features_filtered = df[df['features'].isin(['Single Player', 'Multiplayer', 'Co-op'])]
```
